## Use the sf_temps dataset from vega_datasets for the exercises below:

```python
temps = data.sf_temps()
temps
```

### Resample by the day and take the average temperature. Visualize the average temperature over time.

```python
temps = temps.set_index('date')
temps
```

```python
temps.temp.resample('D').mean()
```

### Write the code necessary to visualize the minimum temperature over time.

```python
temps.temp.resample('D').min().plot()
```

### Write the code necessary to visualize the maximum temperature over time.

```python
temps.temp.resample('D').max().plot()
```

### Which month is the coldest, on average?

```python
temps.temp.resample('M').mean().nsmallest(1).index.month_name()
```

### Which month has the highest average temperature?

```python
temps.temp.resample('M').mean().nlargest(1).index.month_name()
```

```python
temps.temp.resample('M').mean().plot()
```

### Resample by the day and calculate the min and max temp for the day (Hint: `.agg(['min', 'max'])`). Use this resampled dataframe to calculate the change in temperature for the day. Which month has the highest daily temperature variability?


```python
mm_temps = temps.temp.resample('D').agg(['min', 'max'])
mm_temps
```

```python
mm_temps['difference'] = mm_temps['max'] - mm_temps['min']
mm_temps
```

```python
mm_temps.difference.resample('M').max().nlargest(1).index.month_name()
```

### Bonus: Visualize the daily min, average, and max temperature over time on a single line plot, i.e. the min, average, and maximum temperature should be 3 seperate lines.

```python
mean_temps = temps.temp.resample('D').mean()
min_temps = temps.temp.resample('D').min()
max_temps = temps.temp.resample('D').max()

fig, ax = plt.subplots(figsize=(16,9))

ax.plot(max_temps, label='Max Daily Temp', c='g')
ax.plot(mean_temps, label='Mean Daily Temp', c='b')
ax.plot(min_temps, label='Min Daily Temp', c='r')
ax.legend(fontsize=12)
plt.ylabel('Temperature (\N{DEGREE SIGN}F)', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('Max, Min, and Mean Daily Temps for San Francisco in 2010', fontsize=16)
myFmt = DateFormatter("%B")
ax.xaxis.set_major_formatter(myFmt)
```
